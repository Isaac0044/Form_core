```python
# @title **define class "get_form_web" 擷取網站


import requests
import time
import os
from seleniumwire import webdriver
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys




class get_form_web():
  def __init__(self,url):
    self.url=url
    self.headers = {
      'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) Applewebkit/537.36 (KHTML, like Gecko) Chrome/113.0.0.0 Safari/537.36'
    }

    chrome_options = Options()
    chrome_options.add_argument('--no-sandbox')
    chrome_options.add_argument('--headless')
    chrome_options.add_argument('--disable-dev-shm-usage')

    self.wb = webdriver.Chrome(options=chrome_options)
    self.wb.implicitly_wait(2)

    # time.sleep(2)


  def get_web_response(self):

    number, Q_type, Q, Q_option = self.__get_web()
    # 測試變數區
    # for i in range(len(number)):
    #   print("---------------------")
    #   print("目前題號：", f"{i} ")
    #   print("目前題目：", number[i])
    #   print("目前題型：", Q_type[i])
    #   print("目前選項：\n", Q_option[i])
    #   print("-----------------------")
    return number, Q_type, Q, Q_option

  # def not_empty(s):
  #   return s and s.strip()


  def __get_web(self):
      # wb.implicitly_wait(2)
      wb = self.wb
      url = self.url
      headers = self.headers

      wb.get(url)
      wb.implicitly_wait(3)
      # time.sleep(2)

      # 初始化最終的結果列表
      # global number,Q_type,Q,Q_option

      Q_type = []
      number = []
      Q = []
      Q_option = []

      try:
        # time.sleep(1)
        wb.implicitly_wait(1)
        iph_bottom = wb.find_element(By.CLASS_NAME,'iph-button')
        iph_bottom.click()
        print("已經進入頁面，點擊確認")
      except Exception as e:
        time.sleep(1)

      i = 1
      # 遍歷所有的 div
      divs = wb.find_elements(By.XPATH, '//div[@jsname="b0t70b"]')
      # 測試div
      print("之後要隱藏起來：(session) /n",divs)

      for div in divs:
          question_elemt = None
          question = None

      # 選項
        # 取得所有選項文字
          # 在當前 div 下查找所有具有 data-initial-value 的元素
          elements = div.find_elements(By.XPATH, './/*[@data-initial-value]')

          # 提取這些元素的 data-initial-value 性屬值並打包到一個列表中
          option = [elem.get_attribute('data-initial-value') for elem in elements]

          # 將這個列表加入到最終結果中
          options = filter(lambda s: s and s.strip(), option)

          options = list(options)

        # 檢查 '其他' 選項
          # time.sleep(1)  # 等待
          try:
            other_option = div.find_element(By.CLASS_NAME, "PHs6qd")  # 使用相對 CLASS_NAME
            if other_option.text != "":  # 確保文本比較正確
              options.append("其他：*-*-*")
              print(f"===其他選項觸發===\n")
          except Exception as e:
            # print(f"*其他選項*區塊異常: {e}")
            other_option = None

      #題目
          # 查找問題
          try:
              question = div.find_element(By.CLASS_NAME, "KRoqRc")  # 使用相對 CLASS_NAME 限制在當前 div
              question = question.text
          except Exception as e:
              print(f"*問題*區塊異常: {e}")
              question = None

      #題型
          try:
          # 查找 aria-selected="true" 和 tabindex="0" 的元素
            question_type_element = div.find_element(By.XPATH, './/*[@aria-selected="true" and @tabindex="0"]')
            question_type_check = question_type_element.get_attribute('data-value')

            match question_type_check:
              case "0":
                # 簡答
                question_type = "填空"
              case "1":
                # 詳答
                question_type = "填空"
              case "2":
                # 選擇題
                question_type = "單選。"
              case "4":
                # 核取方塊
                question_type = "(可複選)"
              case "5":
                # 線性刻度
                question_type = "五分"
                five_options = []
                options_top_down =[]
                # question_type_element.click()

                # print(div.find_element(By.CLASS_NAME, "zStNOd").is_displayed())
                options_top_down_list = div.find_elements(By.CLASS_NAME, "zStNOd")


                # # 提取這些元素的 data-initial-value 性屬值並打包到一個列表中
                options_top_down_text = [int(options_top_down.get_attribute("textContent")) for options_top_down in options_top_down_list]

                for x in range((options_top_down_text[0]),((options_top_down_text[-1])+1)):
                  five_options.append((x))
                if options == []:
                  options = five_options
                  options = [str(z) for z in five_options]
                else:
                  del five_options[0]
                  five_options.insert(0,options[0])
                  del five_options[-1]
                  five_options.append(options[-1])
                  options = [str(z) for z in five_options]

                # print("options：：",(options))


              case _:
                # 7 單選方格 \ -1 核取方塊格 \ 9 日期 \ 10 時間 \ 檔案上傳 \ 3 下拉式選單
                # 預設填空 全部移至預設 意見區
                question_type = "填空"



          except Exception as e:
            # print(f"*題型*區塊異常: {e}")
            question_type = ""



          # 當辨識不到類型時，清空
          if question_type != "":
            number.append(f"{i}。")
            # 題目
            Q.append(question)
            # 題型
            Q_type.append(question_type)
            # 選項
            Q_option.append(options)

            print("目前題號：",f"{i}。")
            print("目前題目：",question)
            print("目前題型：",question_type)
            print("目前選項：\n",options)
            print("----------------------------------------------")
            i+=1

      return number,Q_type,Q,Q_option
      wb.quit()  # 退出瀏覽器


if __name__ == "__main__":


  url = URL
  get_form = get_form_web(url)
  number, Q_type, Q, Q_option = get_form.get_web_response()

  print("*完成*")
```

```python
import pandas as pd
# import fitz  # PyMuPDF
from collections import deque
# from pypinyin import lazy_pinyin
import re

# EXCEL
import openpyxl
from openpyxl.styles import Font
from openpyxl.utils import get_column_letter
from openpyxl import Workbook
from openpyxl.styles import numbers, NamedStyle, Font, Border, Side, PatternFill, Alignment, Protection
from openpyxl.chart import PieChart, Reference,PieChart3D
from openpyxl.chart.label import DataLabelList
from openpyxl.drawing.text import RegularTextRun, CharacterProperties, Paragraph, ParagraphProperties, Font
from tqdm import tqdm # 設定進度條總長
from datetime import datetime

import os
import unicodedata
import time as time


class process_excel():
  def __init__(self,xlsx_file_name,number,Q_type,Q,Q_option):
    # 原始excel

    self.xlsx_file_name = xlsx_file_name

    self.excel = openpyxl.load_workbook(xlsx_file_name)
    self.excel_sheet = self.excel.worksheets[0]
    self.first_sheet = "表單回應 1"
    self.analysis_sheet = '分析表'
    self.Opinion_sheet = "意見表"
    self.chart_sheet = '圖表'
    self.analysis_all_sheet = '分析總表'
    self.number,self.Q_type,self.Q,self.Q_option = number,Q_type,Q,Q_option

    # 可改為自行修改版本
    file_path = "" # @param {type:"string"}
    ## @param {type:"string"}
    if file_path == "":
      current_time = datetime.now()
      formatted_time = current_time.strftime("20%y_%m_%d")
      # 印出目前的時間
      file_path += f"{xlsx_file_name[:-5]}_表單分析_{formatted_time}"
    file_path += '.xlsx'

    self.file_path = file_path

    row_count = self.read_excel()

    print("環境設置 ==> 完成\n")

  #---------------工具副程式-------------

  # 數字位置轉換為英文位置
  def ct (self,current_column):
    column_text = get_column_letter(current_column)
    return column_text

  # 讀取excel 取得選項數量
  def read_excel(self):
    # 全文掃描
    # 取得特定列 col
    df = pd.read_excel(self.xlsx_file_name)
    column_index = 0  # 選擇目標索引 從0開始計數
    column_data = df.iloc[:, column_index]

    # 計算選項數量(非空值，不包含題目)
    row_count = column_data.count()
    print("\n選項數量:", row_count)
    print()

    # 篩選出其他選項 第一維度:位置(原始) 第二維度 目標選項
    self.row_count = row_count
    self.df = df
    self.other_list = [[],[]]
    return row_count

  def get_form_list(self):
    return self.number,self.Q_type,self.Q,self.Q_option

  def download_excel(self):
    self.excel.save(self.file_path)

    # self.excel.close

    print(f"\n\n  已成功分析完成 ==> {self.file_path}")
    time.sleep(0.5)
    print(f"\n\n 即將為您下載\n")
    time.sleep(1)
    files.download(self.file_path)
    time.sleep(1)
    print(f"\n\n 下載完成\n")

  #---------------工具副程式-------------

  # Step 03-0 建立所有資料表
  def create_all_sheet(self):
    self.analysis_sheet_creat = self.excel.create_sheet(self.analysis_sheet)
    self.Opinion_sheet_creat = self.excel.create_sheet(self.Opinion_sheet)
    self.chart_sheet_create = self.excel.create_sheet(self.chart_sheet)
    self.analysis_all_sheet_create = self.excel.create_sheet(self.analysis_all_sheet)


  # Step 03-1 VBA_Write_分析表----------------------------------------------------

  def an_sheet(self,spacing,sheet_title,run_opinion):
    number,Q_type,Q,Q_option = self.get_form_list()
    print("VBA_Write_分析表 -> running")
    other_list = self.other_list

    # 開啟分析表
    worksheet = self.excel[sheet_title]
    # 圖表所需分析表位置對應陣列
    excel_position_current_row = []
    excel_position_current_column = []

    ##位置設定觀念：  current_row >> 1、2、3...  current_column >> A、B、C...

    # 紀錄五分位置
    five_postion = []
    # 確認是否存在五分
    five_one = False

    # 設定下個其他選項位置(其他題)
    next_other_position = 0

    #基本參數
    Opinion_count =1

    # 1. 設定題目初始位置固定為 -> B2
    current_column = 2  # 列編號B對應2 => B
    current_row = 2   #2

    progress_bar = tqdm(total=len(number*4))
    print()


    for i in range(len(number)):
      progress_bar.update(1)

      #<題目> OK
      #1.設定迴圈讀取所有題目(Q)，並逐一執行填入 2.跳過設定間隔到下一題 3.篩選出填空題型並跳過
      if Q_type[i] != "填空":
        worksheet.cell(row=current_row, column=current_column, value=Q[i])
        # 圖表位置對應陣列 -> 題目
        excel_position_current_row.append(current_row)
        excel_position_current_column.append(current_column)
        current_column += spacing  # 向右移動格子

        # <意見表> OK
      # 1.建立"意見"資料表  2.將填空題型整行移動到意見資料表中
      elif Q_type[i] == "填空" and run_opinion == True:
        #到表單回應1讀取整列資料
        To_first_sheet = self.excel[self.first_sheet]
        first_col_position = self.ct(i+2);Get_first_sheet_col_value = To_first_sheet[first_col_position]
        read_first_sheet_col = To_first_sheet[first_col_position]

        for idx, cell in enumerate(read_first_sheet_col, start=1):
          To_Opinion_sheet = self.excel[self.Opinion_sheet];self.Opinion_sheet_creat.cell(row=idx, column=Opinion_count, value=cell.value)
        Opinion_count += 1
        # 回到分析表
        worksheet = self.excel[sheet_title]

      else:
        print("警告：有超過限定範圍以外之題型(單選、五分、多選、填空)")
        continue

    print(f"  意見表 ==> 完成 /n -------- /n Opinion_sheet is finished ")
    print(f"\n  分析表-題目 ==> 完成")


    #-------------------------------------------------選項--------------------------------------------------
    # 1. 設定題目初始位置固定為 -> A3
    current_column = 1  # 列編號B對應1 => A
    current_row = 3   #3

    for row_data in range(len(number)):
        progress_bar.update(1)

      # 目前五分先以基礎數值為標準，之後以表單方式詢問要以何種形式文字表示之五分量表各項量
        # if Q_type[row_data] == "五分":
        #   mo = ['非常同意','同意','尚可','不同意','非常不同意',]

        #   for cell_data in mo:
        #     worksheet.cell(row=current_row, column=current_column, value=cell_data)
        #     current_row += 1
        # elif Q_type[row_data] != "填空":
        if Q_type[row_data] != "填空":
          for cell_data in Q_option[row_data]:
            if cell_data != "其他：*-*-*":
              worksheet.cell(row=current_row, column=current_column, value=cell_data)
              # 圖表位置對應陣列 -> 選項(填空)
              excel_position_current_row.append(current_row)
              excel_position_current_column.append(current_column)
              current_row += 1
            else:
              worksheet.cell(row=current_row, column=current_column, value="其他：")
              # 圖表位置對應陣列 -> 選項(填空)
              excel_position_current_row.append(current_row)
              excel_position_current_column.append(current_column)
              current_row += 1
              # 標註目前為 其他：選項
              # print(f"其他選項以觸發  當下位置current_column{current_column}  標駐點  {other_list[0][0]-1} 確切轉換公式{2+((other_list[0][0]-1)*spacing)}")



        if Q_type[row_data] != "填空":
          worksheet.cell(row=current_row, column=current_column, value="合計")
          # 圖表位置對應陣列 -> 選項(填空)
          excel_position_current_column.append(current_column)
          excel_position_current_row.append(current_row)
          current_row = 3  # 重置到第三行
          current_column += spacing  #移動到下一題

    print(f"  分析表-選項 ==> 完成")



    #-------------------------------------------------數值(數字)--------------------------------------------------
    # 1. 設定題目初始位置固定為 -> B3
    current_column = 2  # 列編號B對應2 => B
    current_row = 3  #3
    five_one = False #判斷是否有出現過五分
    for row_data in range(len(number)):
        progress_bar.update(1)

        # print("選項:", Q_option[row_data])
        if Q_type[row_data] == "五分":
    #------------------------分析總表 蒐集路徑資訊------------------------
          five_one = True;five_postion.append(row_data)
    # --------------------------------------------------------------------
          ## Q_option[row_data] = ['5','4','3','2','1']

          # # @markdown *-*-*-*-*-*-*-*-*-*-*-*-*
          # # @markdown 五分量表選項(選填)
          # Five_Points_Scale_5 = "非常同意" # @param {type:"string"}
          # Five_Points_Scale_4 = "同意" # @param {type:"string"}
          # Five_Points_Scale_3 = "普通" # @param {type:"string"}
          # Five_Points_Scale_2 = "不同意" # @param {type:"string"}
          # Five_Points_Scale_1 = "非常不同意" # @param {type:"string"}

          # Q_option[row_data] = [Five_Points_Scale_5,Five_Points_Scale_4,Five_Points_Scale_3,Five_Points_Scale_2,Five_Points_Scale_1]

          for cell_data in range (1,len(Q_option[row_data])+1):
              value="=COUNTIF('表單回應 1'!" + str(self.ct(row_data+2)) + ":" + str(self.ct(row_data+2)) + "," + '"' + str(cell_data) + '")'
              worksheet.cell(row=current_row, column=current_column, value=value)
              # 圖表位置對應陣列 -> 選項(五分)
              excel_position_current_row.append(current_row)
              excel_position_current_column.append(current_column)

              current_row += 1

          value="=SUM(" + str(self.ct(current_column)) + "3:" + str(self.ct(current_column)) + str(current_row-1) + ')'
          worksheet.cell(row=current_row, column=current_column, value=value)
          # 圖表位置對應陣列 -> 數值
          excel_position_current_row.append(current_row)
          excel_position_current_column.append(current_column)
        elif Q_type[row_data] == "單選。":
          # 確認目前選項是否有"其他"選項
          now_is_other_bol = False

          for cell_data in Q_option[row_data]:
            # 處理 "其他" 選項------------ (單選其他選項 完成)
              value_connect = ""
              # 04-20 修復 避免空陣列尋找
              if other_list[0] != []:

                if  current_column == (2+((other_list[0][next_other_position]-1)*spacing)) and cell_data == "其他：*-*-*":
                  # print(2+((other_list[0][0]-1)*spacing))

                  value_connect = "="
                  for other_option in range (len(other_list[0])):
                    # =COUNTIF('表單回應 1'!D:D,"小一拉拉")+COUNTIF('表單回應 1'!D:D,"小二")
                    if current_column == (2+((other_list[0][other_option]-1)*spacing)):
                      # 當其他選項超過1後使用 + 將兩個選項連接
                      if now_is_other_bol == True and cell_data == "其他：*-*-*":

                        value_connect += "+"

                      # print(other_list[0][other_option])
                      # 列印對應的value
                      value_connect += "COUNTIF('表單回應 1'!" + str(self.ct(row_data+2)) + ":" + str(self.ct(row_data+2)) + "," + '"' + (other_list[1][other_option][0]) + '"' + ')'
                      # 還原其他選項布林
                      # print(value_connect)

                      now_is_other_bol = True

                      if other_option+1 < len(other_list[0]):
                        if (2+((other_list[0][other_option+1]-1)*spacing)) != current_column:
                          next_other_position = other_option+1

                          # print(next_other_position)
                          # print(other_list[1][other_option][0])
                          # print(other_list[0][other_option])
                          break;

              if value_connect == "":
                value="=COUNTIF('表單回應 1'!" + str(self.ct(row_data+2)) + ":" + str(self.ct(row_data+2)) + "," + str(self.ct(current_column-1)) + str(current_row) + ')'
              else:
                value=value_connect

              # 將自動生成的2個空格改為1個空格
              # cell_data = cell_data.replace("  ", " ")

              worksheet.cell(row=current_row, column=current_column, value=value)
              # 圖表位置對應陣列 -> 數值
              excel_position_current_row.append(current_row)
              excel_position_current_column.append(current_column)
              current_row += 1



          value="=SUM(" + str(self.ct(current_column)) + "3:" + str(self.ct(current_column)) + str(current_row-1) + ')'
          worksheet.cell(row=current_row, column=current_column, value=value)
          # 圖表位置對應陣列 -> 數值
          excel_position_current_row.append(current_row)
          excel_position_current_column.append(current_column)
        elif Q_type[row_data] == "(可複選)":
          # 確認目前選項是否有"其他"選項
          now_is_other_bol = False

          for cell_data in Q_option[row_data]:

            # 處理 "其他" 選項------------ (多選其他選項 )
              value_connect = ""
              # 04-20 修復 避免空陣列尋找
              if other_list[0] != []:
                if  current_column == (2+((other_list[0][next_other_position]-1)*spacing)) and cell_data == "其他：*-*-*":

                  # print(2+((other_list[0][0]-1)*spacing))
                  value_connect = "="
                  # for other_option in range (len(other_list[0])):
                  #   # =COUNTIF('表單回應 1'!D:D,"小一拉拉")+COUNTIF('表單回應 1'!D:D,"小二")
                  #   if current_column == (2+((other_list[0][other_option]-1)*spacing)):
                  #     # 當其他選項超過1後使用 + 將兩個選項連接
                  #     if now_is_other_bol == True and cell_data == "其他：*-*-*":
                  #       value_connect += "+"

                  #     # print(other_list[0][other_option])
                  #     # 列印對應的value
                  #     # 舊版公式
                  #     # value_connect += "COUNTIF('表單回應 1'!"+ str(ct(row_data+2)) + ":" + str(ct(row_data+2)) +',CONCATENATE(,"*",'+ '"' + (other_list[1][other_option][0]) + '"' + "," + '"*"'+"))"
                  #     # =SUMPRODUCT(LEN('表單回應 1'!S:S)-LEN(SUBSTITUTE('表單回應 1'!S:S,",","")))+COUNTA('表單回應 1'!S:S)-1-SUM(CT3:CT10)
                  #     # 新版公式 (公式邏輯： 因為每個版多選回覆中都以","進行分割，故計算","出現的次數，加上第一個出現的值，得出所有的回答次數，再扣掉固定回答，得出其他回答) !! 弊端為若回答中有", "則可能會失準!!
                  #     # value_connect += "COUNTIF('表單回應 1'!"+ str(ct(row_data+2)) + ":" + str(ct(row_data+2)) +',CONCATENATE(,"*",'+ '"' + (other_list[1][other_option][0]) + '"' + "," + '"*"'+"))"
                  #     value_connect += "SUMPRODUCT(LEN('表單回應 1'!"+ str(ct(row_data+2)) + ":" + str(ct(row_data+2)) + ")-LEN(SUBSTITUTE('表單回應 1'!"+ str(ct(row_data+2)) + ":" + str(ct(row_data+2)) + ',",","")))' + "+COUNTA('表單回應 1'!" + str(ct(row_data+2)) + ":" + str(ct(row_data+2)) + ")-1-SUM(" + str(ct(current_column)) + "3:" + str(ct(current_column)) + str(current_row-1) + ')'
                  value_connect += "SUMPRODUCT(LEN('表單回應 1'!"+ str(self.ct(row_data+2)) + ":" + str(self.ct(row_data+2)) + ")-LEN(SUBSTITUTE('表單回應 1'!"+ str(self.ct(row_data+2)) + ":" + str(self.ct(row_data+2)) + ',",","")))' + "+COUNTA('表單回應 1'!" + str(self.ct(row_data+2)) + ":" + str(self.ct(row_data+2)) + ")-1-SUM(" + str(self.ct(current_column)) + "3:" + str(self.ct(current_column)) + str(current_row-1) + ')'

                  #     # value_connect += "COUNTIF('表單回應 1'!" + str(ct(row_data+2)) + ":" + str(ct(row_data+2)) + "," + '"' + (other_list[1][other_option][0]) + '"' + ')'
                  #     # 還原其他選項布林

                  #     print(value_connect)

                  #     now_is_other_bol = True

                  #     if other_option+1 < len(other_list[0]):
                  #       if (2+((other_list[0][other_option+1]-1)*spacing)) != current_column:
                  #         next_other_position = other_option+1
                  #         # print(next_other_position)
                  #         # print(other_list[1][other_option][0])
                  #         # print(other_list[0][other_option])
                  #         break;


              # 將自動生成的2個空格改為1個空格
              # cell_data = cell_data.replace("  ", " ")

              if value_connect == "":
                value="=COUNTIF('表單回應 1'!"+ str(self.ct(row_data+2)) + ":" + str(self.ct(row_data+2)) +',CONCATENATE(,"*",'+ str(self.ct(current_column-1)) + str(current_row) + "," + '"*"'+"))"
              else:
                value=value_connect

                # print(other_list)
                # del other_list[0][0]
                # del other_list[1][0]
                # print(other_list)
              # 將自動生成的2個空格改為1個空格
              # cell_data = cell_data.replace("  ", " ")

              # value="=COUNTIF('表單回應 1'!"+ str(ct(row_data+2)) + ":" + str(ct(row_data+2)) +',CONCATENATE(,"*",'+ str(ct(current_column-1)) + str(current_row) + "," + '"*"'+"))"

              # print(value)
              worksheet.cell(row=current_row, column=current_column, value=value)
              # 圖表位置對應陣列 -> 數值
              excel_position_current_row.append(current_row)
              excel_position_current_column.append(current_column)
              current_row += 1


          # =COUNTIF('表單回應 1'!M:M,CONCATENATE(,"*",BO11,"*"))
          value="=SUM(" + str(self.ct(current_column)) + "3:" + str(self.ct(current_column)) + str(current_row-1) + ')'
          worksheet.cell(row=current_row, column=current_column, value=value)
          # 圖表位置對應陣列 -> 數值
          excel_position_current_row.append(current_row)
          excel_position_current_column.append(current_column)


          # if str(al_value(current_column,current_row)) < str(cell_count()):
          #   # print(f"{row_data+2}選項總和: {al_value(current_column,current_row)}")
          #   # print(f"總回答數量: {cell_count()-1}")
          #   # print(str(ct(row_data+2)))
          #   for i in unique_data(row_data+2,current_row):
          #     print(i)
          #     print(len(i))

          current_row -= len(Q_option[row_data])

        else:
          #! print("填空?")
          continue

        if Q_type[row_data] != "填空":
          current_row = 3  # 重置到第三行
          current_column += spacing  # 移動到下一題
    print(f"  分析表-數值(數字) ==> 完成\n")

    #-------------------------------------------------百分比--------------------------------------------------
    # 1. 設定題目初始位置固定為 -> C3
    current_column = 3  # 列編號B對應3 => C
    current_row = 3  # 3

    for row_data in range(len(number)):
      progress_bar.update(1)

      if Q_type[row_data] != "填空":
        for cell_data in Q_option[row_data]:
            cell_data = cell_data.replace("  ", " ")
            # 計算每個選項數值百分比公式
            value = "="+str(self.ct(current_column-1))+str(current_row)+"/"+str(self.ct(current_column-1))+"$"+str(len(Q_option[row_data])+3)
            cell = worksheet.cell(row=current_row, column=current_column, value=value)
            # 設定百分比格式
            cell.number_format = numbers.FORMAT_PERCENTAGE
            # 圖表位置對應陣列 -> 百分比
            excel_position_current_row.append(current_row)
            excel_position_current_column.append(current_column)

            current_row += 1  # 移動到下一題
        # 計算全部選項數值百分比公式
        value="=SUM(" + str(self.ct(current_column)) + "3:" + str(self.ct(current_column)) + str(current_row-1) + ')'
        cell = worksheet.cell(row=current_row, column=current_column, value=value)
        # 設定百分比格式
        cell.number_format = numbers.FORMAT_PERCENTAGE
        # 圖表位置對應陣列 -> 百分比
        excel_position_current_row.append(current_row)
        excel_position_current_column.append(current_column)
        current_row = 3  # 重置到第三行
        current_column += spacing  # 向右移動

    # 完成後關閉進度條
    progress_bar.close()
    # excel.save(self.file_path)

    print(f"-------- /n {sheet_title} is finished")
    self.other_list = other_list
    self.five_postion = five_postion
    self.five_one = five_one
    self.excel_position_current_row = excel_position_current_row
    self.excel_position_current_column = excel_position_current_column

  def an_chart_process(self,an_spacing,chart_spacing):

    number,Q_type,Q,Q_option = self.get_form_list()
    # 分析表處理(同步處理意見表)
    self.an_sheet(an_spacing,self.analysis_sheet,1)
    # 圖表分析表區塊 處理
    self.chart_spacing = chart_spacing
    self.an_sheet(self.chart_spacing,self.chart_sheet,0)


  # Step 03-3 VBA_Write_分析總表(不可與圖表、分析表互換位置)----------------------------------------------------
  def ana_sheet(self):
    five_postion = self.five_postion
    five_one = self.five_one
    # 取得表單陣列
    number,Q_type,Q,Q_option = self.get_form_list()
    analysis_all_sheet = self.analysis_all_sheet
    print("VBA_Write_分析總表")
    progress_bar = tqdm(total=len(five_postion))
    print()

    # 移動到"分析總表"中
    worksheet = self.excel[analysis_all_sheet]

    # 讀取"圖表"所需的"分析表"位置
    #-------------------------------------------------分析總表(標題列)--------------------------------------------------
    # 1. 設定最上方標題列位置固定為 -> A1
    current_column = 1  # 列編號A對應1 => A
    current_row = 1  # 1

    # 寫入最上方標題列
    if five_one == True:
      cell = worksheet.cell(row=current_row, column=current_column, value="題目");current_column +=1
      for i in range(len(Q_option[(five_postion[0])])):
        cell = worksheet.cell(row=current_row, column=current_column, value=Q_option[(five_postion[0])][i])
        current_column +=1
      # 寫入尾部資料標籤
      cell = worksheet.cell(row=current_row, column=current_column, value="總和")
      current_column +=1
      cell = worksheet.cell(row=current_row, column=current_column, value="分數")
      current_column +=1
      cell = worksheet.cell(row=current_row, column=current_column, value="平均值")
      current_column +=1
      print(f"  總表標題 ==> 完成")


    #-------------------------------------------------分析總表(數據)--------------------------------------------------
    # 1. 設定引入&計算數據位置固定為 -> A2
    current_column = 1  # 列編號A對應1 => A
    current_row += 1   # 2

    if five_one == True:
      for i in range(len(five_postion)):
        # 計算加權分數公式(動態式 支援五分量表以外) 中間數
        count_sum = "="

        # *可加入序號 ="1-1"&D1  ='表單回應 1'!A1
        value = "='表單回應 1'!"+ str(self.ct(five_postion[i]+2))+"1"
        # print(value)
        # 輸入題目
        cell = worksheet.cell(row=current_row, column=current_column, value=value)
        current_column += 1
        # 取的題目為幾維量表
        Q_op_count = len(Q_option[five_postion[i]])
        # 輸入選項公式
        for z in range(len(Q_option[five_postion[i]])):
          progress_bar.update(1)

          # *可加入序號 ="1-1"&D1
          value="=COUNTIF('表單回應 1'!" + str(self.ct(five_postion[i]+2)) + ":" + str(self.ct(five_postion[i]+2)) + "," + '"' + str(int(Q_op_count)-z) + '")'
          cell = worksheet.cell(row=current_row, column=current_column, value=value)
          # 計算加權分數公式(引入參數)
          count_sum += f"{str(self.ct(current_column))}{str(current_row)}*{str(int(Q_op_count)-z)}+"
          current_column += 1
        count_sum += "0"

        # 計算總和 =SUM(B4:F4)
        value="=SUM(" + str(self.ct(current_column-len(Q_option[five_postion[i]]))) + str(current_row) + ":" + str(self.ct(current_column -1 )) + str(current_row)  + ')'
        cell = worksheet.cell(row=current_row, column=current_column, value=value)
        worksheet[str(self.ct(current_column))+str((current_row))].number_format = '0.00'
        current_column += 1
        # 計算加權分數 =B2*5+C2*4+D2*3+E2*2+F2
        cell = worksheet.cell(row=current_row, column=current_column, value=count_sum)
        current_column += 1
        # 計算平均值公式
        value="="+str(self.ct(current_column-1))+ str((current_row))+ "/" + str(self.ct(current_column-2))+ str((current_row))
        cell = worksheet.cell(row=current_row, column=current_column, value=value)

        # 將樣式應用於單元格，設置數字格式為小數點後兩位
        worksheet[str(self.ct(current_column))+str((current_row))].number_format = '0.00'
        #! print(f"小數點處理結果： 儲存格{worksheet[str(ct(current_column))+str((current_row))].value}")


        current_column =1
        current_row +=1
        count_sum = ""
        # print(f"運行{ct(five_postion[i]+2)}行 已成功")

    print(f"  總表數據 ==> 完成")

  # Step c03-2 VBA_Write_圖表----------------------------------------------------
  def chart_sheet_draw(self,Pie_style):
    number,Q_type,Q,Q_option = self.get_form_list()
    chart_sheet = self.chart_sheet
    spacing = self.chart_spacing
    excel_position_current_row = self.excel_position_current_row
    excel_position_current_column = self.excel_position_current_column
    other_list = self.other_list

    print("\nVBA_Write_圖表")
    progress_bar = tqdm(total=len(number)+len(excel_position_current_column))
    print()

    worksheet = self.excel[chart_sheet]
    z =0


    # 計算最多選項
    max_value = max(excel_position_current_row)
    # 計算總共題數
    count_Q = excel_position_current_row.count(2)
    n_sp = 0
    # worksheet = wb[chart_sheet]

    for i in range(len(number)):
      progress_bar.update(1)
      # 畫一個圓餅圖
      if Q_type[i] == "填空":
        n_sp += 1

      else:
            # @markdown 圖片標題大小
        chart_title_font_size = 14 # @param {type:"integer"}
        # labels = Reference(worksheet, min_col=(1+(i*spacing)), min_row=3, max_row=len(Q_option[i])+2) # 標示文字
        # data = Reference(worksheet, min_col=(2+(i*spacing)), min_row=2, max_row=len(Q_option[i])+2)   # 資料
        labels = Reference(worksheet, min_col=(1+((i-n_sp)*spacing)), min_row=3, max_row=len(Q_option[(i)])+2) # 標示文字
        data = Reference(worksheet, min_col=(2+((i-n_sp)*spacing)), min_row=2, max_row=len(Q_option[(i)])+2)   # 資料


        # 建立圓餅圖

        if Pie_style == "PieChart3D":

          pie = PieChart3D()


        elif Pie_style == "PieChart":
          pie = PieChart()
          # 圓餅圖標籤、資料匯入
        pie.add_data(data, titles_from_data=True)
        pie.set_categories(labels)
        pie.legend.position = "b"
        # 顯示類別名稱、百分比與指引線
        dLbls = DataLabelList()
        # 標籤值(百分比)
        dLbls.showPercent = True
        # 標籤值
        dLbls.showVal = False
        # 標籤名
        dLbls.showCatName = True
        dLbls.showSerName = False
        dLbls.showLegendKey = False
        # 標籤指引線
        dLbls.showLeaderLines = True
        # @markdown
        # @markdown 圖片標籤大小
        # @markdown
        # 標籤大小設置為10
        dLbls.size = 10 # @param {type:"integer"}

        # 將數據標籤放在外側
        dLbls.position = "bestFit"
        pie.dLbls = dLbls

        # 圓餅圖標題與大小
        pie.title = Q[i]
        # @markdown

        # @markdown
        # 圓餅圖大小與字體

        title_font = CharacterProperties(ea = Font(typeface = u'標楷體'),sz=chart_title_font_size*100,b=False)
        pie.title.tx.rich.p[0].pPr.defRPr = title_font

        from openpyxl.chart.text import RichText

        font = Font(typeface=u'標楷體')
        D_size = dLbls.size*100 # 10 point size
        cp = CharacterProperties(ea = Font(typeface = u'標楷體'),sz=D_size,b=False)




        # cp = CharacterProperties(latin=font, sz=size, b=False) # Try bold text
        pp = ParagraphProperties(defRPr=cp)
        rtp = RichText(p=[Paragraph(pPr=pp, endParaRPr=cp)])

        pie.legend.textProperties = rtp
        T_font = Font(typeface='Times New Roman')
        T_cp = CharacterProperties(ea = Font(typeface = 'Times New Roman'),sz=D_size,b=False) # Try bold text
        T_pp = ParagraphProperties(defRPr=T_cp)
        T_rtp = RichText(p=[Paragraph(pPr=T_pp, endParaRPr=T_cp)])
        pie.dataLabels.textProperties = T_rtp

        pie.dLbls.textProperties = rtp

        # 設定圓餅圖寬度
        pie.width = (spacing-1)*2

        # # 設置標籤字體
        # for LB in pie.dLbls:
        #   LB.tx.rich.p[0].pPr.defRPr = CharacterProperties(ea = Font(typeface = u'標楷體'))
        #   LB.tx.rich.p[1].pPr.defRPr = CharacterProperties(ea = Font(typeface = u'Times New Roman'))


          # LB.tx.rich.p[0].r.rPr.lang = "zh-CN"  # 中文
          # LB.tx.rich.p[0].r.rPr.sz = 200        # 中文字體大小為10pt
          # LB.tx.rich.p[0].r.rPr.typeface = u'標楷體'  # 中文字體
          # LB.tx.rich.p[1].r.rPr.lang = "en-US"  # 英文
          # LB.tx.rich.p[1].r.rPr.sz = 200        # 英文字體大小為10pt
          # LB.tx.rich.p[1].r.rPr.typeface = "Times New Roman"  # 英文字體




        # 將圖形放置在最底下的儲存格位置(在空一格)
        worksheet.add_chart(pie, f"{self.ct(1+((i-n_sp)*spacing))}{max_value+1}")






if __name__ == "__main__":
  # form data list => number, Q_type, Q, Q_option

  run_excel = process_excel(xlsx_file_name,number,Q_type,Q,Q_option)
  run_excel.read_excel()
  run_excel.create_all_sheet()

  an_spacing = 4
  chart_spacing = 8 # @param {type:"integer"}
  # 處理 分析表 圖表 意見表
  run_excel.an_chart_process(an_spacing,chart_spacing)

  # 處理分析總表
  run_excel.ana_sheet()

  # 繪製圖表
  Pie_style = "PieChart" # @param ["PieChart3D", "PieChart"]
  run_excel.chart_sheet_draw(Pie_style)

  # 下載
  run_excel.download_excel()








```
